# Cleanning
### To avoid losing the original data, I will make a copy for each and work on these copies.

```python
t_archive_cleans=t_archive.copy()
i_pred_clean=i_pred.copy()
tweet_df_cleans=tweet_df.copy()
```

##### Define: Drop the columns with too much missing values in  `t_archive` and `tweet_df`  and unneeded columns in the analysis, using drop and filter functions.

##### Code:

```python
#Drop in t_archive table
t_archive_cleans=t_archive_cleans.drop(['in_reply_to_user_id', 'in_reply_to_status_id', 'retweeted_status_id', 'retweeted_status_user_id', 'retweeted_status_timestamp'], axis=1)
```

##### Test:

```python
t_archive_cleans.info()
```

##### Code:

```python
#Use filtier rathar than drop to select only the needed columns.
tweet_df_cleans=tweet_df_cleans.filter(['id','favorite_count','retweet_count', 'source', 'user'])
```

##### Test:

```python
tweet_df_cleans.info()
```

##### Define: Convert the type to datetime usin astype function.

##### Code:

```python
t_archive_cleans['timestamp'] = t_archive_cleans['timestamp'].astype('datetime64')
```

##### Test:

```python
t_archive_cleans.head(2)
```

As appear the +0000 problem solved also when the type changed 

##### Define: In order to unification the denominator, any value that not equal 10 will be drop.

##### Code:

```python
t_archive_cleans.drop(t_archive_cleans[t_archive_cleans.rating_denominator != 10].index, inplace=True)
```

##### Test:

```python
t_archive_cleans.rating_denominator.value_counts()
```

##### Define: for the rating_numerator, most values under 20, any value is bigger will drop

##### Code:

```python
t_archive_cleans.drop(t_archive_cleans[t_archive_cleans.rating_numerator >= 20].index, inplace=True)
```

```python
t_archive_cleans.drop(t_archive_cleans[t_archive_cleans.rating_numerator == 0].index, inplace=True)
```

##### Test:

```python
t_archive_cleans.rating_numerator.value_counts()
```

##### Define:replace none with nan to cosider as null values using replace function.

##### Code:

```python
t_archive_cleans =  t_archive_cleans.replace('None', np.nan)
```

##### Test:

```python
t_archive_cleans.info()
```

##### Define: drop some invalid names using drop function.

##### Code:

```python
actually=t_archive_cleans[t_archive_cleans['name']=='actually']
a=t_archive_cleans[t_archive_cleans['name']=='a']
an=t_archive_cleans[t_archive_cleans['name']=='an']
the=t_archive_cleans[t_archive_cleans['name']=='the']
t_archive_cleans.drop(actually.index,inplace=True)
t_archive_cleans.drop(a.index,inplace=True)
t_archive_cleans.drop(an.index,inplace=True)
t_archive_cleans.drop(the.index,inplace=True)
```

##### Test:

```python
t_archive_cleans[t_archive_cleans['name']=='an']
```

```python
t_archive_cleans[t_archive_cleans['name']=='the']
```

##### Define: Combine the four columns to one column(dog_type) using loc function. 

##### Code:

```python
t_archive_cleans.loc[t_archive_cleans['doggo'] == 'doggo', 'dog_type'] = 'doggo'
t_archive_cleans.loc[t_archive_cleans['floofer'] == 'floofer', 'dog_type'] = 'floofer'
t_archive_cleans.loc[t_archive_cleans['pupper'] == 'pupper', 'dog_type'] = 'pupper'
t_archive_cleans.loc[t_archive_cleans['puppo'] == 'puppo', 'dog_type'] = 'puppo'
```

# Analysis and visualization

In this phase, I reached the results for the following questions:-
1.	Retweeted tweets are favorites and vice versa? 
2.	What are the most famous dogs type?
3.	What is the top dog breed?
4.	What the most favorite dog breed?


```python
twitter_archive_master=pd.read_csv('twitter_archive_master.csv')
twitter_archive_master.head(2)
```

### Retweeted tweets are favorite and vice verse?

```python
twitter_archive_master.plot(kind = 'scatter', x = 'favorite_count', y = 'retweet_count', alpha = 0.5, color = 'blue')
plt.ylabel('Retweets')
plt.xlabel('Favorites')
plt.title('Retweets and favorites Tweets');
```

### What is the most famous dogsâ€™ type?

```python
twitter_archive_master['dog_type'].value_counts().plot(kind='bar',color='green')
plt.title('The dog types')
plt.xlabel('Dog type')
plt.ylabel('Count');
```

### What is the top dog breed?

```python
twitter_archive_master.groupby('probability_breeds').count()['rating_numerator'].sort_values(ascending=False).nlargest(12).plot(kind='bar',figsize=(10,7))
plt.ylabel("Dog Breed",fontsize=14)
plt.xlabel("Tweet count",fontsize=14)
plt.title("The top 12 dogs breed",fontsize=18);
```

### What the most favorite dog breed?

```python
df1 = round(twitter_archive_master.groupby('probability_breeds')['favorite_count'].sum() / 
                  twitter_archive_master.groupby('probability_breeds')['favorite_count'].count(),0)
```

```python
#top 30  favorite dog breed
df1.sort_values(ascending=False).nlargest(30).plot(kind='bar',color='red',figsize=(15,8))
plt.title("Top 30 Average of favorite dog breed",fontsize=18)
plt.ylabel("Favorite averge ")
plt.xlabel("Dog Breed");
```
